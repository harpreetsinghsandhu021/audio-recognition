## Splitting the Dataset

```python
# The audio clips are 1 second or less at 16kHz. The output_sequence_length=16000 pads the short 
# ones to exactly 1 second (and would trim longer ones) so that they can be easily batched.
train_ds, val_ds = tf.keras.utils.audio_dataset_from_directory(
    directory=data_dir, 
    batch_size=64, 
    validation_split=0.2, 
    seed=0, 
    output_sequence_length=16000,
    subset='both'
)

label_names = np.array(train_ds.class_names)
label_names
```

The dataset now contains batches of audio clips and integer labels. The audio clips have a shape of (batch, samples, channels).

```python
train_ds.element_spec
```

This dataset only contains single channel audio, so use the tf.squeeze function to drop the extra axis

```python
def squeeze(audio, labels): 
    audio = tf.squeeze(audio, axis=-1)
    return audio, labels
```

```python
train_ds = train_ds.map(squeeze, tf.data.AUTOTUNE)
val_ds = val_ds.map(squeeze, tf.data.AUTOTUNE)
```

```python
train_ds.element_spec
```

The `utils.audio_dataset_from_directory` function only returns up to two splits. It's a good idea to keep a test set separate from your validation set. Ideally you'd keep it in a separate directory, but in this case you can use `Dataset.shard` to split the validation set into two halves. Note that iterating over any shard will load all the data, and only keep its fraction.

```python
# Split the validation dataset into 2 parts
# index=0 means take the first shard
test_ds = val_ds.shard(num_shards=2, index=0)

# Take the second shard of the validation dataset
# index=1 means take the second shard 
val_ds = val_ds.shard(num_shards=2, index=1)
```

### Convert Waveforms to Spectograms

The waveforms in the dataset are represented in the time domain. Next, you'll transform the waveforms from the time-domain signals into the time-frequency-domain signals by computing the `short-time Fourier transform (STFT)` to convert the waveforms to as spectrograms, which show frequency changes over time and can be represented as 2D images. You will feed the spectrogram images into your neural network to train the model. 

### Short-time Fourier Transform 
The STFT essentially creates a time-frequency "image" that shows how the frequency content of the signal changes over time, making it much more informative than either pure time-domain or pure frequency-domain representations.

```python
# This Function converts an audio waveform into a spectogram representation that can be used as 
# input for CNN Models
def get_spectogram(waveform): 
    # Apply STFT to convert tieme-domain signals into frequency domain
    # frame_length=255 indicates that each window contains 255 samples 
    # frame_step=128 indicates that windows overlap by moving forward 128 samples each time
    spectogram = tf.signal.stft(waveform, frame_length=255, frame_step=128)

    # Calculate absolute values of the complex STFT 
    # this gives the us the strength output of frequencies at each time point 
    spectogram = tf.abs(spectogram)

    # Add an extra dimension, this creates an extra dimension required for CNN input
    # Final shape will be: [batch_size, time_steps, frequencies, 1]
    spectogram = spectogram[..., tf.newaxis]

    return spectogram
```

```python
# This Function assists in displaying a spectrogram

def plot_spectrogram(spectrogram, ax): 
    if len(spectrogram.shape) > 2:
        # verfy it`s exactly three dimensions
        assert len(spectrogram.shape) == 3

        # remove single dimensional entries from array shape
        spectrogram = np.squeeze(spectrogram, axis=-1)

    # convert spectogram to logarithmic scale and transpose 
    # `spectogram.T` transposes the spectogram matrix, swapping rows and columns. this is done to put time on x-axis 
    # `np.finfo(float).eps` is the smallest positive number that can be represented by the float type(approx. 2.22e-16),we
    # add this number because: 
    # np.log(0) is undefined(returns negative infinty) 
    # spectograms often contain zero values 
    # adding this tiny number prevents log(0) errors while not significantly affecting the results
    # `np.log()` takes the natural log of all values because log scaling makes it easier to visualize both loud and quiet components
    log_spec = np.log(spectrogram.T + np.finfo(float).eps)

    # Get dimensions of the log spectogram 
    height = log_spec.shape[0] # frequency steps
    width = log_spec.shape[1] # time steps

    # create x-axis points evenly spaces from 0 to spectogram size
    X = np.linspace(0, np.size(spectrogram), num=width, dtype=int)

    # create y-axis point from 0 to height 
    Y=range(height)

    ax.pcolormesh(X, Y, log_spec)
```

Now, Create Spectogram datasets from audio datasets

```python
def process_audio(audio, label):
    return get_spectogram(audio), label

def make_spectrogram_dataset(ds): 
    return ds.map(process_audio, num_parallel_calls=tf.data.AUTOTUNE)
```

```python
train_spectrogram_ds = make_spectrogram_dataset(train_ds)
test_spectrogram_ds = make_spectrogram_dataset(test_ds)
val_spectrogram_ds = make_spectrogram_dataset(val_ds)
```

### Model

```python
train_spectrogram_ds = train_spectrogram_ds.cache().shuffle(10000).prefetch(tf.data.AUTOTUNE)
val_spectrogram_ds = val_spectrogram_ds.cache().prefetch(tf.data.AUTOTUNE)
test_spectrogram_ds = test_spectrogram_ds.cache().prefetch(tf.data.AUTOTUNE)
```

```python
input_shape = example_spectrograms.shape[1:]
num_labels = len(label_names)

norm_layer = tf.keras.layers.Normalization()

# adapt method to calculate stats(mean and standard deviation)
norm_layer.adapt(data=train_spectrogram_ds.map(map_func=lambda spec, label : spec))

model = models.Sequential([
    layers.Input(shape=input_shape),
    layers.Resizing(32, 32), 
    norm_layer, 
    layers.Conv2D(32, 3, activation='relu'),
    layers.Conv2D(64, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Dropout(0.25), 
    layers.Flatten(), 
    layers.Dense(128, activation='relu'), 
    layers.Dropout(0.5), 
    layers.Dense(num_labels)
])
```

```python
model.summary()
```

### Compile the Model

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(), 
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), 
    metrics=['accuracy']
)
```

```python

history = model.fit(
    train_spectrogram_ds,
    validation_data=val_spectrogram_ds,
    epochs=10,
    # callbacks=tf.keras.callbacks.EarlyStopping(verbose=1, patience=2),
)
```

```python
metrics = history.history
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.plot(history.epoch, metrics['loss'], metrics['val_loss'])
plt.legend(['loss', 'val_loss'])
plt.ylim([0, max(plt.ylim())])
plt.xlabel('Epoch')
plt.ylabel('Loss [CrossEntropy]')

plt.subplot(1,2,2)
plt.plot(history.epoch, 100*np.array(metrics['accuracy']), 100*np.array(metrics['val_accuracy']))
plt.legend(['accuracy', 'val_accuracy'])
plt.ylim([0, 100])
plt.xlabel('Epoch')
plt.ylabel('Accuracy [%]')
```
